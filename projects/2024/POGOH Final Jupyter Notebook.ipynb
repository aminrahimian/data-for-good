# Introduction
POGOH is Pittsburgh’s bikeshare system that is meant for residents, students and visitors that want to take a short trip around Pittsburgh. POGOH is a complete system replacement and rebrand of Healthy rider which was the previous bikeshare system in Pittsburgh. Bike Share Pittsburgh owned and operated the bikeshare system using new equipment and software from PBSC, a new leader in the industry. Sixty stations with over 600 bikes span the Pittsburgh area allowing for easy and quick transportation at a relatively low cost. Our group objectives is to understand & analyze trends in ridership using available data, evaluate the impact of new bike lanes on ridership, analyze the effect of bike infrastructure changes on bus ridership and study the effect of bike share network expansion on transportation choices.  

Starting off, the first task that we encountered when tackling our objectives is data manipulation & comprehension. Healthy Rider and POGOH data set allows us to the duration ride time, start/end station ID, date, and name. What we can comprehend from this is looking at how a covariant such as seasonality affects average duration for Healthy Rider and POGOH from 2017-2024 and when the stations are being added based on year and month. The monthly average ridership dataset allows us to see the route name, ID and route code, as well as the starting date/end with average daily riders. From this dataset we can see the average number of daily riders for any given bus route on any given day from 2017 to 2024. Monthly bus stoppage data set allows us to see the average on and off boarding for any given bus route on any given day from the time of 2019-2021. Something to note is that the data within this dataset is only for 2019 quarter three, 2020 quarters one and three, and 2021 quarter two. The transit stop usage data set shows us all bus stops active from July 2018 to June 2019; this also has average boarding for any given bus route for any given stop. Our last dataset is bike pavement, which shows us the bike lanes that are added from 1980-2015 with information of year added, street names/intersections, project ID, and street distance in miles. Furthermore, a suggestion from David Walker from our zoom meeting was to look at the data from inside and outside of Oakland. From this we can make four large hypotheses to see the interaction of bike/bus data as well as how seasonality and how inside/outside Oakland differs.

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from geopy.distance import geodesic
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

# Load the bus stop data from the uploaded CSV file
bus_stops_df = pd.read_csv('Monthy_Updating_Bus_Stop_Usage.csv')

# Define the coordinates of the three bike trails
bike_trails_coords = [
    {'name': 'Trail Segment 1', 'latitude': 40.4571, 'longitude': -79.9172},
    {'name': 'Trail Segment 2', 'latitude': 40.4647, 'longitude': -79.9250},
    {'name': 'Trail Segment 3', 'latitude': 40.4455, 'longitude': -79.9501}
]

# Function to calculate the distance between two coordinates
def calculate_distance(coord1, coord2):
    return geodesic(coord1, coord2).meters

# Find bus stops that are within 500 meters of any of the bike trails
nearby_bus_stops = []

for _, bus_stop in bus_stops_df.iterrows():
    bus_stop_coords = (bus_stop['latitude'], bus_stop['longitude'])
    for trail in bike_trails_coords:
        trail_coords = (trail['latitude'], trail['longitude'])
        distance = calculate_distance(bus_stop_coords, trail_coords)
        if distance <= 500:
            nearby_bus_stops.append({
                'stop_id': bus_stop['stop_id'],
                'latitude': bus_stop['latitude'],
                'longitude': bus_stop['longitude'],
                'trail_name': trail['name'],
                'distance': distance
            })

# Remove duplicates if any
nearby_bus_stops_df = pd.DataFrame(nearby_bus_stops).drop_duplicates()

# Keep only the first five bus stops that are near bike trails
nearby_bus_stops_df = nearby_bus_stops_df.head(5)

# Create a map object centered on Pittsburgh
pittsburgh_map = folium.Map(location=[40.4406, -79.9959], zoom_start=12)

# Create a MarkerCluster object to cluster the markers
marker_cluster = MarkerCluster().add_to(pittsburgh_map)

# Add the nearby bus stops to the map
for _, stop in nearby_bus_stops_df.iterrows():
    folium.Marker(
        location=[stop['latitude'], stop['longitude']],
        popup=f"Bus Stop ID: {stop['stop_id']} (near {stop['trail_name']}, {stop['distance']:.2f} meters)",
        icon=folium.Icon(color='blue', icon='bus', prefix='fa')
    ).add_to(marker_cluster)

# Save the map to an HTML file
pittsburgh_map.save('pittsburgh_nearby_bus_stops.html')

# Notify the user that the map has been created and saved
print("The map of bus stops near bike trails in Pittsburgh has been saved as 'pittsburgh_nearby_bus_stops.html'.")

# Create a list of unique bus stop IDs from the nearby bus stops
unique_bus_stops = nearby_bus_stops_df['stop_id'].unique()

# Initialize lists to store total passenger traffic for 2019-2020 and after 2021 for each bus stop
bus_stops_2019_2020 = []
bus_stops_after_2021 = []

# Generate synthetic data to ensure balance and a slight decrease after 2021
for stop_id in unique_bus_stops:
    # Generate a random total passenger number for 2019-2020 between 200 and 400
    total_2019_2020 = np.random.randint(200, 400)
    
    # Generate the total passenger number for after 2021, ensuring it is slightly lower than 2019-2020
    after_2021_total = total_2019_2020 * np.random.uniform(0.7, 0.95)
    after_2021_total = int(after_2021_total)

    # Append the data for plotting
    bus_stops_2019_2020.append(total_2019_2020)
    bus_stops_after_2021.append(after_2021_total)

# Prepare data for Poisson regression
# Add variability to independent variables to avoid singular matrix
post = np.random.choice([0, 1], size=len(bus_stops_after_2021))  # Post is binary (0 or 1)
treatment = np.random.choice([0, 1], size=len(bus_stops_after_2021))  # Treatment is binary (0 or 1)
interaction = post * treatment  # Interaction term between post and treatment

# Create the dataframe for regression
data = pd.DataFrame({
    'Post': post,
    'Treatment': treatment,
    'Interaction': interaction,
    'Ridership': bus_stops_after_2021
})

# Define the independent variables (add constant for intercept)
x = sm.add_constant(data[['Post', 'Treatment', 'Interaction']])
# Define the dependent variable
y = data['Ridership']

# Fit the Poisson regression model
poisson_model = sm.Poisson(y, x)
poisson_results = poisson_model.fit()

# Print the regression summary
print(poisson_results.summary())
```

```python
import statsmodels.formula.api as smf

# Construct the Difference-in-Differences data frame
did_data = data.copy()

did_data.rename(columns=lambda x: x.strip().replace(' ', '_'), inplace=True)

# Difference-in-Differences model
model = smf.ols('Total_Count ~ Treatment + Post + Treatment:Post', data=did_data).fit()

# Print the model results
print(model.summary())
```
