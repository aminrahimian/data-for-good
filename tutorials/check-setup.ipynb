## Use Jupyter Notebook to Run the codes

To install Jupyter Notebook in your computer, it is recommended to use Anaconda.

You can install it here: https://www.anaconda.com/products/distribution#Downloads

After you install the Anaconda, you can see that jupyter notebook is already installed there. Two ways to run it:
1. Open the Anaconda Navigator, find the Jupyter Notebook and launch it
2. Open the Anaconda Prompt, type "jupyter notebook", then press enter to run it.

Jupyter Notebook has ipykernel already there for us to run the python code. 

It is a little bit difficult to run R in jupyter notebook. It is recommended to use R studio instead.

If you want to run R code in Jupyter Notebook, you must first install IRkernel.
Here is a tutorial on it: https://datatofish.com/r-jupyter-notebook/

To run both Python and R in one jupyter notebook, it is necessary to install rpy2 library.
Steps:
1. Install rpy2 by typing: "pip install rpy2"
2. Add enviornment variables
    - R_HOME: The root path of R. For example: C:\Program Files\R\R-4.1.2\
    - R_USER: C:\Users\Administrator（The default user name）
    
    In the PATH, add:
    
    - %R_HOME%\bin\x64\R.dll（64 bits）
    - %R_HOME%\bin\x64
3. Check in Jupyter:
    - !python -m rpy2.situation
    - %load_ext rpy2.ipython (Each time before you run R code, you should run this one first)
4. Usage
    - To install package:
        - %R install.packages('ggplot2')
    - To interact between python and r:
        - Cell 1:
        
        import pandas as pd
        
        df = pd.DataFrame({'x':[1,2,3],'y':[1,32,3]})
        
        - Cell 2:
        
        %%R -i df -o data
        
        data=df
        
        "%%" means the cell is in R code, -i means input from python and -o means output
